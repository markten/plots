# Temperature Dependence of LED Color

A change in an LED's color may be induced by a large temperature change. This color change is related to the way in which thermal energy impacts the bandgap of the material used to construct the LED.

The bandgap of a material at a set temperature may be given approximately by Varshni's eqation.

\begin{align}
    E_{g}(T) = E_{g}(0)-\frac{\alpha T^2}{T+\beta}
\end{align}

In this equation, $E_{g}(0)$ is the material's bandgap at 0 Kelvin. $\alpha$ and $\beta$ are fitted parameters.

Once the bandgap is calculated, it may be translated into a wavelength via the Planck-Einstein relation.

\begin{align}
    E = \frac{hc}{\lambda}
\end{align}

Here $E$ is the energy associated with the bandgap, $h$ is Planck's constant in terms of eV, $c$ is the speed of light, and $\lambda$ is the wavelength.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.colors as color

cmin = 380
cmax = 700

E_g0 = {}
alpha = {}
beta = {}

# from http://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.488.9625&rep=rep1&type=pdf

E_g0['germanium'] = 0.744
E_g0['silicon'] = 1.116
E_g0['gaas'] = 1.519
E_g0['gap'] = 2.09
E_g0['ingan'] = 0.69
E_g0['znse'] = 2.8250

alpha['germanium'] = 4.07*10**-4
alpha['silicon'] = 4.73*10**-4
alpha['gaas'] = 5.41*10**-4
alpha['gap'] = 4.8*10**-4
alpha['ingan'] = 4.1*10**-4
alpha['znse'] = 5.32*10**-4

beta['germanium'] = 230
beta['silicon'] = 636
beta['gaas'] = 204
beta['gap'] = 358
beta['ingan'] = 454
beta['znse'] = 190

T = np.linspace(0,400,1000)

def calc_lambda(key):
    h = 4.135667662*10**-15
    c = 3*10**8
    return (h*c) / (E_g0[key] - ((alpha[key]*T**2)/(T+beta[key]))) *10**9
```

```python
lam = calc_lambda('gap')

plt.scatter(T, lam, c=lam, cmap=cm.nipy_spectral, edgecolor='none', vmin=cmin, vmax=cmax)
plt.xlabel('Temperature [K]')
plt.ylabel('Wavelength')
plt.title('GaP')

plt.show()
```

```python
lam = calc_lambda('gaas')

plt.scatter(T, lam, c=lam, cmap=cm.nipy_spectral, edgecolor='none', vmin=cmin, vmax=cmax)
plt.xlabel('Temperature [K]')
plt.ylabel('Wavelength')
plt.title('GaAs')

plt.show()
```

```python
lam = calc_lambda('germanium')

plt.scatter(T, lam, c=lam, cmap=cm.nipy_spectral, edgecolor='none', vmin=cmin, vmax=cmax)
plt.xlabel('Temperature [K]')
plt.ylabel('Wavelength')
plt.title('Germanium')

plt.show()
```

```python
lam = calc_lambda('ingan')

plt.scatter(T, lam, c=lam, cmap=cm.nipy_spectral, edgecolor='none', vmin=cmin, vmax=cmax)
plt.xlabel('Temperature [K]')
plt.ylabel('Wavelength')
plt.title('InGaN')

plt.show()
```

```python
lam = calc_lambda('znse')

plt.scatter(T, lam, c=lam, cmap=cm.nipy_spectral, edgecolor='none', vmin=380, vmax=760)
plt.xlabel('Temperature [K]')
plt.ylabel('Wavelength')
plt.title('ZnSe')

plt.show()
```
